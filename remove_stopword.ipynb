```python
count={}
for j in range(len(a)):
  for i in a[j]:
      try: count[i] += 1
      except: count[i]=1
print(count)
```

```python
search=['및','등','업','용','류','내','방','위','도','물','미만','재','캐'] #불용어 추가하면 됨.
for i in range(len(search)):
  for j in range(len(a)):
    for item in a[j]:
      if item ==search[i]:
          a[j].remove(search[i])
```
