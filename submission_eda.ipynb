```python
col = ['text_obj','text_mthd','text_deal']
NLP = submission[col].apply(lambda row: ' '.join(row.values.astype(str)), axis=1)
NLP.head(100)
```

```python
import re
def text_cleaning(text):
    hangul = re.compile('[^ ㄱ-ㅣ가-힣]+') # 한글의 정규표현식을 나타냅니다.
    result = hangul.sub('', text)
    return result
```
